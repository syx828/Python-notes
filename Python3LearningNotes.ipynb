```python
import sys #使用sys模块中的getrefcount函数
import weakref 
class Animal(object):
    def run(self):
        print('%s is running...' %(self.name))
class Dog(Animal):
    def __init__(self):
        self.name = 'Dog'
    def eat(self):
        print('Dog is eating meat...')
class Rabbit(Animal):
    def __init__(self):
        self.name = 'Rabbit'
    def eat(self):
        print('Rabbit is eating carrot...')
dog = Dog()
print(dog.__weakref__)
d = dog
print(dog.__weakref__)
d = None
(1,'普通引用不是弱引用')
print(1,'生成一个引用时，引用计数加一：',sys.getrefcount(dog))
Dp = weakref.proxy(dog)
print()
print(2,'添加一个代理对象时，引用计数不增加：',sys.getrefcount(dog),'dog的弱引用对象为：',dog.__weakref__)
DD = weakref.ref(dog)
print(3,'创建一个弱引用，实质上是生成一个新的对象：',sys.getrefcount(dog),'dog的弱引用对象变为：',dog.__weakref__)
dd = DD()
print(4,'弱引用的实例，也是弱引用对象的一个实例，创建弱引用的实例会使弱引用对象的引用计数增加：',sys.getrefcount(dog))
print('删除对象的全部实例后，Python会删除该对象')
print('（只删掉删掉弱引用的对象dog）')
dog = None
print('DD run:',end=' ')
DD().run()
print(5,'此时对象还存在，因为弱引用的实例还存在，该对象的引用计数还不为0')
print('（再删掉弱引用的实例）')
print('（此时对象已消失）')
dd = None
try:
    print('DD run:',end=' ')
    DD().run()
except AttributeError:
    print()
    print(6,'弱引用失效，导致一个AttributeError，因为DD()现在是NoneType')
try:
    print('Dp run：',end=' ')
    Dp.run()
except ReferenceError:
    print()
    print(7,'代理失效，将会导致一个ReferenceError，因为Dp现在指向的对象不存在')

a = 3
b = 3
x = 1.0
print()
print('8 0的引用次数为%s, 1的引用次数为%s, 2的引用次数为%s, 3的引用次数为%s, 1.0的引用次数为%s' %(sys.getrefcount(0),sys.getrefcount(1),sys.getrefcount(2),sys.getrefcount(3),sys.getrefcount(1.0)))
```

```python
import math
def quadratic(a, b, c):
    #参数检查
    for i in (a,b,c):
        if not isinstance(i,(int,float)):
            raise ValueError('Invalid Input!')
    if a==0:
        return -c/b
    else:
        delta = b*b-4*a*c
        if delta==0:
            return -b/a/2
        elif delta>0:
            delta = math.sqrt(delta)
            return tuple([(-b+delta)/a/2,(-b-delta)/a/2])
        else:
            delta = math.sqrt(-delta)
            return tuple([(-b+delta*1j)/a/2,(-b-delta*1j)/a/2])
# 测试:
print('quadratic(2, 3, 1) =', quadratic(2, 3, 1))
print('quadratic(1, 3, -4) =', quadratic(1, 3, -4))
print('quadratic(2, 2, 1) =', quadratic(2, 2, 1))
print('quadratic(0, 2, 1) =', quadratic(0, 2, 1))

if (quadratic(2, 3, 1) != (-0.5, -1.0)) & (quadratic(2, 3, 1) != (-1.0, -0.5)):
    print('测试失败')
elif (quadratic(1, 3, -4) != (1.0, -4.0)) & (quadratic(1, 3, -4) != (-4.0, 1.0)):
    print('测试失败')
elif (quadratic(2, 2, 1) != ((-0.5+0.5j), (-0.5-0.5j))) & (quadratic(2, 2, 1) != ((-0.5-0.5j), (-0.5+0.5j))):
    print('测试失败')
elif (quadratic(0, 2, 1) != -0.5):
    print('测试失败')
else:
    print('测试成功')
quadratic(0, '2', 1)
        
```

```python
def fact(n,s=''):
    if n<0:
        raise ValueError('Wrong Input!')
    if s is '':
        print(s+'fact('+str(n)+')')
    else:
        print('='+s+'fact('+str(n)+')')
    if n<1:
        return 1
    t = fact(n-1,s+str(n)+'*')
    r = n*t
    print('='+s+str(r))
    return r
fact(5)
fact(0)
```

```python
def fib(n):
    if n<2:
        return 1
    return fib(n-2)+fib(n-1)
fib(3)
```

```python
def fact(n,prod=1):
    if n<2:
        return prod*n
    return fact(n-1,prod*n)
fact(4)
```
